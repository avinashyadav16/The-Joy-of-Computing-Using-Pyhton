- **BARBELL GRAPH**

<pre>

- **COMPLETE GRAPH**
<pre>
Returns the complete graph K_n with n nodes.
A complete graph on n nodes means that 
all pairs of distinct nodes have an edge connecting them.

Parameters:
n : int or iterable container of nodes
    If n is an integer, nodes are from range(n). If n is a container of nodes, those nodes appear in the graph.
</pre>

```python

G = nx.complete_graph(6)
nx.draw(G)
```

<br><br><br>

- **CYCLE GRAPH**

<pre>
Returns the cycle graph C_n of cyclically connected nodes.
C_n is a path with its two end-nodes connected.

n : int or iterable container of nodes

This is two paths of n nodes, with each pair connected by a single edge.
Node labels are the integers 0 to 2*n - 1.

<pre>

p : float

```python
import networkx as nx
import random
import operator
import matplotlib.pyplot as plt
```

```python

# Drawing the Directed Random Graph marked with the labels:
G = nx.gnp_random_graph(10, 0.5, directed=True)
nx.draw(G, with_labels=True)  # Corrected 'with_labels' typo here
plt.show()

# x is random source node from a list of nodes:
x = random.choice( [i for i in range(G.number_of_nodes())] )

# Initialising the dictionary to store the counter:
dict_counter = {}

# Initialising the dictionary values as Zero(0):
for i in range(G.number_of_nodes()):
    dict_counter[i] = 0

# Incrementing the counter of the source node: 
dict_counter[x] = dict_counter[x] + 1

# Looking the neighbours of x and traversing whole of the graph:
for i in range(1000000):
    list_n = list(G.neighbors(x))
    
    # if x is the sink( i.e. it has no more available neighbours ):
    if( len( list_n )==0 ):
        x = random.choice( [i for i in range(G.number_of_nodes())] )
        dict_counter[x] = dict_counter[x] + 1
        
    # if x is not the sink:
    else:
        # Choosing a node randomly from the list of the neighbour:
        x = random.choice(list_n)
        dict_counter[x] = dict_counter[x] + 1        




# Now, using the page rank method:
page_rank = nx.pagerank(G)

sorted_page_rank = sorted( page_rank.items(), key=operator.itemgetter(1)) 
sorted_random_walk = sorted( dict_counter.items(), key=operator.itemgetter(1)) 

for i in sorted_page_rank:
    print(i)
print()

print(sorted_random_walk)
```

```python
def add_edges():
    # Defining the list of nodes
    # For picking two nodes that will be edges:
    nodes = list(G.nodes())
    
    for source_node in nodes:
        for target_node in nodes:
            
            # Avoiding the loops in the graph:
            if source_node != target_node:
                
                # Following a tossing a coin method:
                # that means generating a random method b/w 0 amd 1
                tossed_coin = random.random()
                
                # If we get less than or equal tp 0.5 means we got HEAD 
                # And in this case we'll add the edge:
                if tossed_coin <= 0.5:
                    G.add_edge(source_node, target_node)
    
    return G


#----------------------------------------------------------------------------


def assign_points(G):
    # Defining the list of nodes
    nodes = list(G.nodes())
    
    point = []
    for each in nodes:
        point.append(100)
    
    return point
   

#----------------------------------------------------------------------------


def distributin_points(G, points):
    # Defining the list of nodes
    nodes = list(G.nodes())
    
    new_points = []
    for i in range(len(nodes)):
        new_points.append(0)
        
    for n in nodes:
        out = list(G.out_edges(n))
        if len(out) == 0:
            new_points[n] = new_points[n] + points[n]
        
        else:
            share = points[n] / len(out)
            for (src, tgt) in out:
                new_points[tgt] = new_points[tgt] + share
    
    return new_points


#----------------------------------------------------------------------------


def keep_distributing(points, G):
    # Defining the list of nodes
    nodes = list(G.nodes())
    new_points = None
    
    while(1):
        new_points = distributin_points(G, points)
        print(new_points)
        points = new_points
        stop = input("Press # to stop or any key to continue... ")
        if stop == '#':
            break
    
    return new_points


#----------------------------------------------------------------------------

def rank_by_points(final_points):
    d = {}
    for i in range( len(points) ):
        d[i] = points
    
    print(sorted(d.items(), key=lambda f:f[1]))


#----------------------------------------------------------------------------


# Difining the Directecd Graph:
G = nx.DiGraph()

# Adding Nodes to the graph:
G.add_nodes_from( [i for i in range(10)] )

# Adding edges to the graph:
G = add_edges()


#----------------------------------------------------------------------------


# Visulising the graph:
nx.draw(G, with_labels=True)
plt.show()


#----------------------------------------------------------------------------


# Assigning the initial points:
points = assign_points(G)

# Keep distributing points and we'll get final points or conversed points:
final_points = keep_distributing(points, G)

# Rank by points:
rank_by_points(final_points)

# Default networkx function:
result = nx.pagerank(G)
print(sorted(result.items(), key=lambda f:f[1]))

```
